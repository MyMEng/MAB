```python
# Import
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
```

```python
# First step
m1 = 50
v1 = 300
m2 = 70
v2 = 300

s1 = np.sqrt(v1)
s2 = np.sqrt(v2)

x = np.linspace(0,100,100)

plt.figure(0)

vlines( x[55], 0, mlab.normpdf(x,m1,s1)[55], colors='b' )
plt.plot( x,mlab.normpdf(x,m1,s1),     x,mlab.normpdf(x,m2,s2) )

plt.axis('off')

# vlines(x, ymin, ymax, colors='k', linestyles='solid', label='', **kwargs)

vlines( x[68], 0, mlab.normpdf(x,m2,s2)[68], colors='g' )

# plt.axis([ 0, 1, -0.075, -0.035 ])
plt.savefig("sim1.png", dpi=300, pad_inches=0.1)

plt.show()
```
